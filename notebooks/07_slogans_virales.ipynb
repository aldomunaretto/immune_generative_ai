### Prompts de Sistema

Separamos las instrucciones del sistema del prompt del usuario para mayor control.

````python
def generar_con_sistema(system_prompt, user_prompt, model="gpt-4o-mini"):
    start_time = time.time()
    response = openai.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )
    print(f"{time.time() - start_time:.2f} segundos")
    return response.choices[0].message.content
````

````python
system_prompt = """
Eres un experto en marketing viral y director creativo de √©lite. Tu especialidad es crear esl√≥ganes que:
1. Conectan emocionalmente con el p√∫blico
2. Son extremadamente memorables
3. Tienen potencial viral en redes sociales
4. Generan conversaci√≥n y engagement
5. Se convierten en parte de la cultura popular

Un eslogan exitoso debe incluir:
- M√°ximo 6 palabras
- Elemento emocional fuerte
- Juego de palabras o rima (opcional pero preferible)
- Llamada a la acci√≥n impl√≠cita
- Diferenciaci√≥n clara de la competencia

Siempre proporciona el eslogan y una breve explicaci√≥n de por qu√© ser√° viral.
"""

user_prompt = """
Crea un eslogan para una marca de caf√© premium artesanal llamada "Aurora Coffee". 
El p√∫blico objetivo son millennials urbanos de 25-35 a√±os con ingresos medios-altos.
La marca se posiciona como el caf√© que despierta tu mejor versi√≥n cada ma√±ana.
"""

response = generar_con_sistema(system_prompt, user_prompt)
print(response)
````

# ¬øQu√© es DSPy y por qu√© es importante para Context Engineering?

## Introducci√≥n a DSPy

**DSPy** (Declarative Self-improving Python) es un framework desarrollado por Stanford que revoluciona la manera en que trabajamos con modelos de lenguaje grandes (LLMs). A diferencia del prompting tradicional, DSPy nos permite crear sistemas m√°s robustos, modulares y optimizables autom√°ticamente.

## Problemas del Prompting Tradicional

En las secciones anteriores hemos visto el enfoque tradicional de prompting, que tiene varias limitaciones:

### üö´ Limitaciones del Prompting Manual:
- **Fragilidad**: Los prompts son sensibles a peque√±os cambios
- **No escalable**: Dif√≠cil mantener consistency en aplicaciones grandes
- **Trial & Error**: Optimizaci√≥n manual, lenta y subjetiva
- **Dif√≠cil de versionar**: Hard-coded prompts son dif√≠ciles de mantener
- **No transferible**: Los prompts optimizados para un modelo pueden no funcionar en otro

### üí° Ejemplo de Fragilidad:
```python
# Prompt fr√°gil - peque√±os cambios pueden afectar drasticamente el resultado
prompt_v1 = "Crea un eslogan para caf√©"
prompt_v2 = "Crea un eslogan publicitario para caf√©"  # Mejor resultado
prompt_v3 = "Genera un eslogan publicitario para caf√©"  # Posible degradaci√≥n
```

## ¬øC√≥mo DSPy resuelve estos problemas?

### üèóÔ∏è **Arquitectura Modular**
DSPy separa la **l√≥gica del programa** de los **prompts espec√≠ficos**:
- **Signatures**: Definen QU√â queremos (inputs/outputs)
- **Modules**: Definen C√ìMO procesamos (Predict, ChainOfThought, etc.)
- **Optimizers**: Mejoran autom√°ticamente el rendimiento

### üîÑ **Optimizaci√≥n Autom√°tica**
En lugar de ajustar prompts manualmente, DSPy:
- Genera autom√°ticamente prompts optimizados
- Usa few-shot learning inteligente
- Optimiza bas√°ndose en m√©tricas objetivas
- Se adapta a diferentes modelos autom√°ticamente

### üìä **Evaluaci√≥n Sistem√°tica**
DSPy permite:
- Definir m√©tricas claras de √©xito
- Comparar diferentes enfoques objetivamente
- Iterar r√°pidamente sobre mejoras
- Tracking de performance a lo largo del tiempo

## Context Engineering con DSPy

El **Context Engineering** es el arte de estructurar y optimizar el contexto que proporcionamos a los LLMs para obtener mejores resultados. DSPy lleva esto al siguiente nivel:

### üéØ **Context Engineering Tradicional vs DSPy**

| Aspecto | Tradicional | DSPy |
|---------|-------------|------|
| **Prompts** | Hard-coded, manuales | Generados autom√°ticamente |
| **Ejemplos** | Seleccionados manualmente | Optimizados algor√≠tmicamente |
| **Optimizaci√≥n** | Trial & error manual | B√∫squeda autom√°tica |
| **Evaluaci√≥n** | Subjetiva | M√©tricas objetivas |
| **Mantenimiento** | Alto esfuerzo | Auto-mejorable |

### üöÄ **Beneficios para nuestro caso de uso (Esl√≥ganes Virales)**

1. **Consistency**: Todos los esl√≥ganes seguir√°n la misma estructura de calidad
2. **Optimizaci√≥n**: DSPy puede encontrar los mejores ejemplos few-shot autom√°ticamente
3. **Adaptabilidad**: F√°cil adaptar a diferentes tipos de productos/marcas
4. **Escalabilidad**: Puede manejar m√∫ltiples variaciones simult√°neamente
5. **Mejora continua**: El sistema puede mejorar con m√°s datos y feedback

## Tipos de Modules en DSPy

### üîÆ **dspy.Predict**
- M√≥dulo b√°sico para predicciones simples
- Ideal para tareas directas como nuestros esl√≥ganes
- Genera prompts optimizados autom√°ticamente

### üß† **dspy.ChainOfThought**
- A√±ade razonamiento paso a paso
- Mejora la calidad explicando el "por qu√©"
- Perfecto para decisiones creativas complejas

### üîÑ **dspy.ReAct** 
- Combina razonamiento y acciones
- √ötil cuando necesitamos buscar informaci√≥n adicional

### üéØ **dspy.Retrieve**
- Integra sistemas de b√∫squeda/RAG
- Perfecto para contexto din√°mico basado en conocimiento externo

## ¬øPor qu√© usar DSPy para Marketing y Esl√≥ganes?

### üìà **Casos de uso perfectos para DSPy:**
- **Consistency de marca**: Mantener tono y estilo consistente
- **A/B Testing**: Generar variaciones autom√°ticamente
- **Personalizaci√≥n**: Adaptar mensajes a diferentes segmentos
- **Escalabilidad**: Generar contenido para m√∫ltiples productos/campa√±as
- **Optimizaci√≥n ROI**: Mejorar m√©tricas de engagement autom√°ticamente

### üé® **Lo que veremos en la implementaci√≥n:**
1. **Signature Definition**: C√≥mo definir inputs/outputs para esl√≥ganes
2. **Module Selection**: Cu√°ndo usar Predict vs ChainOfThought
3. **Optimization**: C√≥mo DSPy mejora autom√°ticamente los resultados
4. **Evaluation**: M√©tricas para medir la calidad de esl√≥ganes

---

¬°Ahora veamos DSPy en acci√≥n! üëá

````python
import dspy
dspy.configure(lm=dspy.LM("openai/gpt-4o-mini"))

class EsloganViralSignature(dspy.Signature):
    """
    Eres un experto en marketing viral especializado en crear esl√≥ganes publicitarios 
    que conectan emocionalmente y tienen potencial para volverse virales en redes sociales.
    """
    marca: str = dspy.InputField(description="Nombre de la marca")
    producto: str = dspy.InputField(description="Tipo de producto o servicio")
    publico_objetivo: str = dspy.InputField(description="Descripci√≥n del p√∫blico objetivo")
    posicionamiento: str = dspy.InputField(description="Posicionamiento deseado de la marca")
    
    eslogan_principal: str = dspy.OutputField(description="Eslogan principal (m√°ximo 6 palabras)")
    variacion_1: str = dspy.OutputField(description="Primera variaci√≥n del eslogan")
    variacion_2: str = dspy.OutputField(description="Segunda variaci√≥n del eslogan")
    explicacion_viral: str = dspy.OutputField(description="Explicaci√≥n de por qu√© tiene potencial viral")
    hashtags: str = dspy.OutputField(description="Hashtags sugeridos separados por comas")

generador_eslogan = dspy.Predict(EsloganViralSignature)

resultado = generador_eslogan(
    marca="Aurora Coffee",
    producto="Caf√© premium artesanal",
    publico_objetivo="Millennials urbanos de 25-35 a√±os con ingresos medios-altos",
    posicionamiento="El caf√© que despierta tu mejor versi√≥n cada ma√±ana"
)

print(resultado)
````

## DSPy Predict vs ChainOfThought: ¬øCu√°ndo usar cada uno?

### ü§î **¬øNotaste la diferencia?**

Acabamos de ver la misma `EsloganViralSignature` ejecut√°ndose con dos m√≥dulos diferentes:
1. **`dspy.Predict`**: Resultado directo
2. **`dspy.ChainOfThought`**: Resultado + razonamiento expl√≠cito

### ‚öñÔ∏è **Comparaci√≥n Detallada**

| Aspecto | dspy.Predict | dspy.ChainOfThought |
|---------|--------------|---------------------|
| **Velocidad** | ‚ö° M√°s r√°pido | üêå M√°s lento (genera m√°s tokens) |
| **Costo** | üí∞ M√°s barato | üí∏ M√°s caro (m√°s tokens) |
| **Calidad** | ‚úÖ Buena | üéØ Generalmente mejor |
| **Explicabilidad** | ‚ùå Sin razonamiento | ‚úÖ Razonamiento completo |
| **Debugging** | üîç Dif√≠cil debuggear | üîß F√°cil ver d√≥nde fall√≥ |
| **Confianza** | ü§∑ "Conf√≠a en el resultado" | üí° "Entiendes por qu√©" |

### üéØ **¬øCu√°ndo usar cada uno?**

#### üöÄ **Usa `dspy.Predict` cuando:**
- **Prototipado r√°pido**: Necesitas resultados inmediatos
- **Tareas simples**: El task es directo y bien definido
- **Alto volumen**: Necesitas procesar muchos elementos
- **Presupuesto limitado**: Quieres minimizar costos de API
- **Confianza alta**: Ya validaste que la signature funciona bien

#### üß† **Usa `dspy.ChainOfThought` cuando:**
- **Tareas complejas**: Requiere an√°lisis o creatividad profunda
- **Calidad cr√≠tica**: No puedes permitirte errores
- **Explicabilidad**: Necesitas justificar las decisiones
- **Debugging**: Est√°s optimizando el sistema
- **Presentaciones**: Mostrar√°s el razonamiento a clientes/stakeholders
- **Aprendizaje**: Quieres entender c√≥mo piensa el modelo
